# Group By

```python
# 計算不同廠牌汽車的平均價格，並畫出來
mean_price = data.groupby("make")["price"].mean().reset_index(name="MeanPrice")
ax = sns.barplot(mean_price, x="MeanPrice", y="make")
for index, row in mean_price.iterrows():
    ax.text(row["MeanPrice"] + 1000, # 偏移量，確保數字不會隱形重疊
            index,
            round(row["MeanPrice"], 2), # 顯示價格，保留兩位小數
            color="black",
            ha="left", # 水平對齊
            va="center") # 垂直對齊
plt.show()
```

```python
mean_price = data.groupby("make")["price"].mean().reset_index(name="MeanPrice")
bars = plt.barh(mean_price["make"], mean_price["MeanPrice"], color="skyblue")
# 在每個線形上顯示 MeanPrice 的數值
for bar in bars:
    # 使用 bar.get_width() 來獲取每個線形的長度(即價格)
    width = bar.get_width()
    plt.text(width + 1000,
             bar.get_y() + bar.get_height() / 2, # 垂直位置
             f'{width:2f}', # 顯示價格，保留兩位小數
             color="black",
             va="center") # 垂直對齊
# 顯示圖表
plt.xlabel("Mean Price")
plt.ylabel("Make")
plt.title("Mean Price by Make")
plt.show()
```

```python
# 練習題
# 修改上圖將平均價格由高到低排序
mean_price = data.groupby("make")["price"].mean().reset_index(name="MeanPrice").sort_values(by="MeanPrice", ascending=False)
#mean_price = mean_price.sort_values(by="MeanPrice", ascending=False)
ax = sns.barplot(mean_price, x="MeanPrice", y="make")
for index, row in mean_price.iterrows():
    ax.text(row["MeanPrice"] + 1000, # 偏移量，確保數字不會隱形重疊
            index,
            round(row["MeanPrice"], 2), # 顯示價格，保留兩位小數
            color="black",
            ha="left", # 水平對齊
            va="center") # 垂直對齊
plt.show()
```

```python
mean_price = data.groupby("make")["price"].mean().reset_index(name="MeanPrice").sort_values(by="MeanPrice", ascending=False)
bars = plt.barh(mean_price["make"][::-1], mean_price["MeanPrice"][::-1], color="skyblue")
# 在每個線形上顯示 MeanPrice 的數值
for bar in bars:
    # 使用 bar.get_width() 來獲取每個線形的長度(即價格)
    width = bar.get_width()
    plt.text(width + 1000,
             bar.get_y() + bar.get_height() / 2, # 垂直位置
             f'{width:.2f}', # 顯示價格，保留兩位小數
             color="black",
             va="center") # 垂直對齊
# 顯示圖表
plt.xlabel("Mean Price")
plt.ylabel("Make")
plt.title("Mean Price by Make")
plt.show()
```

```python
# 練習題
# 計算不同車型的平均馬力
mean_horsepower = data.groupby("Body-style")["horsepower"].mean().reset_index(name="MeanHorsepower").sort_values(by="MeanHorsepower", ascending=False)
#mean_price = mean_price.sort_values(by="MeanPrice", ascending=False)
ax = sns.barplot(mean_horsepower, x="MeanHorsepower", y="Body-style")
for index, row in mean_horsepower.iterrows():
    ax.text(row["MeanHorsepower"] + 20, # 偏移量，確保數字不會隱形重疊
            index,
            round(row["MeanHorsepower"], 2), # 顯示價格，保留兩位小數
            color="black",
            ha="left", # 水平對齊
            va="center") # 垂直對齊
plt.show()
```

```python
mean_horsepower = data.groupby("Body-style")["horsepower"].mean().reset_index(name="MeanHorsepower").sort_values(by="MeanHorsepower", ascending=False)
bars = plt.barh(mean_horsepower["Body-style"][::-1], mean_horsepower["MeanHorsepower"][::-1], color="skyblue")
# 在每個線形上顯示 MeanPrice 的數值
for bar in bars:
    # 使用 bar.get_width() 來獲取每個線形的長度(即價格)
    width = bar.get_width()
    plt.text(width + 20,
             bar.get_y() + bar.get_height() / 2, # 垂直位置
             f'{width:.2f}', # 顯示價格，保留兩位小數
             color="black",
             va="center") # 垂直對齊
# 顯示圖表
plt.xlabel("Mean Horsepower")
plt.ylabel("Body-style")
plt.title("Mean Horsepower by Body-style")
plt.show()
```
